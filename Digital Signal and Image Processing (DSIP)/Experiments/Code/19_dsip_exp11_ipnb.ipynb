Sharpening Filter : High Pass Filter

```python
url = "https://eeweb.engineering.nyu.edu/~yao/EL5123/image/barbara_gray.bmp"
image = io.imread(url)
gray_image =image
img = gray_image
#gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
kernel = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])
kernel = kernel/9
hpf1 = cv2.filter2D(img,-1,kernel)
print("High Pass Filter Output")
cv2_imshow(hpf1)
```

```python
kernelhbf = np.array([[-1, -1, -1],
                       [-1, 9, -1],
                       [-1, -1, -1]])
hpf1 = cv2.filter2D(img,-1,kernelhbf)
print("High Boost Filter Output")
cv2_imshow(hpf1)
```
